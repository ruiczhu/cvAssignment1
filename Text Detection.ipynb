```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
```

```python
custom_config = r'--oem 3 --psm 6'
```

```python
#Load the image using cv2
image_path = './test1.jpg'
image = cv2.imread(image_path)
```

```python
def cv2_imshow(image):
    cv2.imshow('image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def reload_image(image_path):
    image = cv2.imread(image_path)
    return image
    
def gbr2rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image = gbr2rgb(image)

plt.imshow(image)
# cv2_imshow(image)
```

```python
#get gray image
def get_grayscale(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return img_gray
```

```python
gray = get_grayscale(image)

plt.imshow(gray)

pytesseract.image_to_string(gray, config=custom_config)
```

```python
def gaussian_blur(image,d=3):
    img_blur3 = cv2.GaussianBlur(image, (2*d+1, 2*d+1), -1)[d:-d,d:-d]
    return cv2.cvtColor(img_blur3, cv2.COLOR_BGR2RGB)
```

```python
blur = gaussian_blur(image, 3)

plt.imshow(blur)

pytesseract.image_to_string(blur, config=custom_config)
```

```python
def thresholding(image):
    img_thresholding = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    return cv2.cvtColor(img_thresholding, cv2.COLOR_BGR2RGB)
```

```python
thresh = thresholding(gray)

plt.imshow(thresh)

pytesseract.image_to_string(thresh, config=custom_config)
```

```python
def canny(image, th1=30, th2=90):
    img_canny = cv2.Canny(image, th1, th2)
    return img_canny
```

```python
img = canny(gray,100,200)

plt.imshow(img)

pytesseract.image_to_string(img, config=custom_config)
```

```python
def sobelx(image):
    img_sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=9)
    sobelx_uint8 = cv2.normalize(img_sobelx, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
    return sobelx_uint8
```

```python
sobel = sobelx(gray)

plt.imshow(sobel, cmap='gray')

pytesseract.image_to_string(sobel, config=custom_config)
```

```python
def sobely(image):
    img_sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=9)
    sobely_uint8 = cv2.normalize(img_sobely, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
    return sobely_uint8
```

```python
sobel = sobely(gray)

plt.imshow(sobel, cmap='gray')

pytesseract.image_to_string(sobel, config=custom_config)
```

```python
def edge_detection(image, th1=30, th2=90, d=3):
    edgeresult = image.copy()
    
    edgeresult = gaussian_blur(edgeresult, d)
    
    gray = get_grayscale(edgeresult)
    
    edge = canny(gray, th1, th2)
    
    edgeresult[edge != 0] = (0, 255, 0)
    
    plt.imshow(cv2.cvtColor(edgeresult, cv2.COLOR_BGR2RGB))
```

```python
edge_detection(image)
```

```python
def detect_char(image):
    detect = image.copy()
    img_h, img_w, _ = detect.shape
    boxes = pytesseract.image_to_boxes(image)
    for n in boxes.splitlines():
        n = n.split(' ')
        x, y, w, h = int(n[1]), int(n[2]), int(n[3]), int(n[4])
        cv2.rectangle(detect, (x, img_h - y), (w, img_h - h),(0,255,0),3)
        cv2.putText(detect, n[0], (x, img_h - y + 25), cv2.FONT_HERSHEY_COMPLEX,1 ,(255,0,0),2)
    result = gbr2rgb(detect)
    plt.imshow(result)

detect_char(image)
```

```python
def detect_words(image):
    detect = image.copy()
    boxes = pytesseract.image_to_data(detect)
    for m,n in enumerate(boxes.splitlines()):
        if m != 0:
            n = n.split()
            if len(n) == 12:
                x, y, w, h = int(n[6]), int(n[7]), int(n[8]), int(n[9])
                cv2.rectangle(detect, (x, y), (w+x, h+y),(0,255,0),3)
                cv2.putText(detect, n[11], (x, y), cv2.FONT_HERSHEY_COMPLEX,1 ,(255,0,0),2)
    result = gbr2rgb(detect)
    plt.imshow(result)

detect_words(image)
```

```python
def detect_num(image):
    detect = image.copy()
    custom_config = r'--oem 3 --psm 6 outputbase digits'
    boxes = pytesseract.image_to_data(detect, config=custom_config)
    for m,n in enumerate(boxes.splitlines()):
        if m != 0:
            n = n.split()
            if len(n) == 12:
                x, y, w, h = int(n[6]), int(n[7]), int(n[8]), int(n[9])
                cv2.rectangle(detect, (x, y), (w+x, h+y),(0,255,0),3)
                cv2.putText(detect, n[11], (x, y), cv2.FONT_HERSHEY_COMPLEX,1 ,(255,0,0),2)
    result = gbr2rgb(detect)
    plt.imshow(result)

detect_num(image)
```
