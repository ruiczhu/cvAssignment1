```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
```

```python
# Define custom configuration for Tesseract OCR
# --oem 3: Use the default OCR Engine Mode (OEM) which is a combination of the standard and LSTM engines
# --psm 6: Set the Page Segmentation Mode (PSM) to assume a single uniform block of text
custom_config = r'--oem 3 --psm 6'
```

```python
image_path = './test1.jpg'
image = cv2.imread(image_path)
```

```python
# Function to convert an image from BGR to RGB color space
def bgr2rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Create a copy of the original image
copy_img = image.copy()

test_image = bgr2rgb(copy_img)

plt.imshow(test_image)
plt.axis('off')
plt.show()
```

```python
# Function to convert an image to grayscale
def get_grayscale(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return img_gray
```

```python
copy_img = image.copy()

gray = get_grayscale(copy_img)

plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()

# Use pytesseract to extract text from the grayscale image
pytesseract.image_to_string(gray, config=custom_config)
```

```python
# Function to apply Gaussian blur to an image
def gaussian_blur(image, d=3):
    img_blur = cv2.GaussianBlur(image, (2*d+1, 2*d+1), -1)[d:-d, d:-d]
    return img_blur
```

```python
copy_img = image.copy()

blur = gaussian_blur(copy_img, 3)

plt.imshow(blur)
plt.axis('off')
plt.show()

# Use pytesseract to extract text from the blurred image
pytesseract.image_to_string(blur, config=custom_config)
```

```python
# Function to apply thresholding to an image
def thresholding(image):
    img_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    return img_thresh
```

```python
copy_img = image.copy()

gray = get_grayscale(copy_img)

thresh_img = thresholding(gray)

plt.imshow(thresh_img, cmap='gray')
plt.axis('off')
plt.show()

# Use pytesseract to extract text from the thresholded image
pytesseract.image_to_string(thresh_img, config=custom_config)
```

```python
# Function to apply Canny edge detection to an image
def canny(image, th1=30, th2=90):
    img_canny = cv2.Canny(image, th1, th2)
    return img_canny
```

```python
copy_img = image.copy()

gray = get_grayscale(copy_img)

canny_img = canny(gray)

plt.imshow(bgr2rgb(canny_img))
plt.axis('off')
plt.show()

pytesseract.image_to_string(canny_img, config=custom_config)
```

```python
def sobelx(image):
    # Compute the Sobel gradient in the x direction
    img_sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=9)
    
    # Normalize the result to the range [0, 255] and convert to uint8
    sobelx_uin8 = cv2.normalize(img_sobelx, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
    
    return sobelx_uin8
```

```python
copy_img = image.copy()

gray = get_grayscale(copy_img)

sobel_x = sobelx(gray)

plt.imshow(sobel_x)
# plt.imshow(bgr2rgb(canny_img))

pytesseract.image_to_string(sobel_x, config=custom_config)
```

```python
def sobely(image):
    img_sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=9)
    sobely_uin8 = cv2.normalize(img_sobely, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
    return sobely_uin8
```

```python
copy_img = image.copy()

gray = get_grayscale(copy_img)

sobel_y = sobely(gray)

plt.imshow(sobel_y)
# plt.imshow(bgr2rgb(canny_img))

pytesseract.image_to_string(sobel_y, config=custom_config)
```

```python
def edge_detection(image, th1=30, th2=90, d=3):
    edge_result = image.copy()  # Create a copy of the original image for processing
    
    edge_result = gaussian_blur(edge_result, d)  # Apply Gaussian blur to reduce noise
    
    gray = get_grayscale(edge_result)  # Convert the blurred image to grayscale
    
    edge = canny(gray, th1, th2)  # Perform Canny edge detection with specified thresholds
    
    edge_result[edge != 0] = (0, 255, 0)  # Highlight detected edges in green
    
    return edge_result  # Return the image with detected edges
```

```python
img_edge = edge_detection(image)

plt.imshow(bgr2rgb(img_edge))

pytesseract.image_to_string(img_edge, config=custom_config)
```

```python
def detect_char(image):
    detect_result = image.copy()  # Create a copy of the original image for character detection
    
    img_h, img_w, _ = detect_result.shape  # Get the height and width of the image
    
    boxes = pytesseract.image_to_boxes(detect_result)  # Use Tesseract to get bounding boxes for detected characters
    
    for n in boxes.splitlines():  # Iterate through each detected character
        n = n.split(' ')  # Split the box information into individual components
        
        x, y, w, h = int(n[1]), int(n[2]), int(n[3]), int(n[4])  # Extract coordinates and dimensions
        
        # Draw a rectangle around the detected character
        cv2.rectangle(detect_result, (x, img_h - y), (w, img_h - h), (0, 255, 0), 3)
        
        # Put the character label above the rectangle
        cv2.putText(detect_result, n[0], (x, img_h - y + 25), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
    
    return detect_result  # Return the image with detected characters highlighted
```

```python
detect_r = detect_char(image)
detect_r = bgr2rgb(detect_r)
plt.imshow(detect_r)
# cv2.imshow('', detect_r)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
def detect_words(image):
    detect_result = image.copy()  # Create a copy of the original image for word detection
    
    img_h, img_w, _ = detect_result.shape  # Get the height and width of the image
    
    boxes = pytesseract.image_to_data(detect_result)  # Use Tesseract to get detailed information about detected words
    
    for m, n in enumerate(boxes.splitlines()):  # Iterate through each line of data
        if m != 0:  # Skip the header line
            n = n.split()  # Split the line into individual components
            
            if len(n) == 12:  # Ensure the line contains the expected number of elements
                x, y, w, h = int(n[6]), int(n[7]), int(n[8]), int(n[9])  # Extract coordinates and dimensions
                
                # Draw a rectangle around the detected word
                cv2.rectangle(detect_result, (x, y), (w + x, h + y), (0, 255, 0), 3)
                
                # Put the word label above the rectangle
                cv2.putText(detect_result, n[11], (x, y), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
    
    return detect_result  # Return the image with detected words highlighted
```

```python
detect_r = detect_words(image)
detect_r = bgr2rgb(detect_r)
plt.imshow(detect_r)
# cv2.imshow('', detect_r)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
